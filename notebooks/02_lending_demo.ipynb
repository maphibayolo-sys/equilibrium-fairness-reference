# Equilibrium Fairness — Lending Demo

**Dataset.** UCI German Credit (synthetic stand-in if no network).
**Metric.** Equalised odds gap on age (under/over 25).
**Threshold.** θ = 0.10.

This notebook demonstrates the same architecture from `01_hiring_demo` operating on a different domain with a different metric. **The architecture is identical; only the configuration changes.** That is the cross-domain claim from §6 of the working paper, made tangible.

```python
import yaml
from pathlib import Path
from src.loop import run

cfg_path = '../configs/lending_german.yaml'
cfg = yaml.safe_load(Path(cfg_path).read_text())

print('Config:')
for k, v in cfg.items():
    print(f'  {k}: {v}')
```

```python
summary = run(cfg_path)
print(summary)
```

## Inspect the audit log

```python
import json
log_lines = Path(summary['log_path']).read_text().strip().split('\n')
for line in log_lines:
    event = json.loads(line)
    flag = '⚠️  BREACH' if event['breach'] else '✓ within'
    print(f"{flag}  window {event['window_start']}-{event['window_end']}  "
          f"|δ|={event['magnitude']:.4f}  θ={event['threshold']}")
```

## What changed between hiring and lending?

Only configuration: dataset name, protected attribute, metric, threshold, window size, κ-owner. The architecture, the modules, and the runtime loop are identical.

This is the framework's structural-universality claim. The political economy of correction in lending is different from in hiring — credit committees operate under different legal constraints from hiring committees — but the *operational architecture* of monitoring and escalation is the same.
